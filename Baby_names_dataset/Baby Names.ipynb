## Group by names.

```python
names_count = baby[['Name', 'Count']]

names_count.groupby('Name').sum().head(10)
```

## Sort by count.

```python
names = names_count.groupby('Name').sum()

names.sort_values('Count', ascending = False).head(10)
```

## Number of unique names.

```python
len(names)
```

```python
#Aliter
#SYNTAX: dataframe.column_name.unique()
uniques = baby.Name.unique()
len(uniques)
```

## Name with the most occurences.

```python
names.sort_values('Count', ascending = False).head(1)
```

```python
#Aliter
names[ names['Count'] == names['Count'].max() ]
```

```python
names_with_min_occ = names[ names['Count'] == names['Count'].min() ]

print(names_with_min_occ.shape)
print('\n')
print(names_with_min_occ.head(10))
```

## Names with median number of occurences.

```python
names[ names['Count'] == names['Count'].median() ]
```
