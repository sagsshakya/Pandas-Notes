```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import copy

# Read in TN middle school dataset from GitHub
df = pd.read_csv('https://raw.githubusercontent.com/LearnDataSci/article-resources/master/Essential%20Statistics/middle_tn_schools.csv')
school = copy.copy(df)
```

```python
school.isnull().sum()
```

```python
sub = school[['reduced_lunch', 'school_rating']]
```

```python
sub.groupby(['school_rating']).describe()
```

```python
sub.corr()
```

```python
plt.figure(figsize = (10,8))
sns.set(style = 'whitegrid', font_scale = 1.2) 
```

## Box - plot for reduce_lunch vs. school rating.

```python
plt.figure(figsize = (10,8))
sns.set(style = 'whitegrid', font_scale = 1.2) 

sns.boxplot(x = 'reduced_lunch', y = 'school_rating', data = school, palette = 'Set2', orient='h')
plt.show()
```

#### Conclusion:
- The higher - rated schools have lower proportions of students in the 'reduced_lunch' scheme. 

## Linear Correlation between 'school_rating' and 'reduced_lunch'.

```python
plt.figure(figsize = (10,8))
sns.set(style = 'whitegrid', font_scale = 1.2) 
sns.regplot(x = school['school_rating'], y= school['reduced_lunch'], scatter= True, marker='o', color = 'b' )
plt.show()
```

#### Conclusion:
- There is a negative correlation between these two variables indicating that when the rating of the school increases, the proportion of students having the 'reduced_lunch' scheme decreases.

## Creating a heatmap (correlation matrix) for checking which variables further affect the school_rating.

```python
school.corr()
```

```python
cor = school.corr().columns.values
print(cor)
print(type(cor))
```

```python
plt.figure(figsize = (10,8))
sns.set(style = 'whitegrid', font_scale = 1.2) 

sns.heatmap(data=school.corr(), xticklabels=cor, yticklabels=cor, cmap = 'coolwarm')
plt.show()

'''
Description of the figure:
- Red cells indicate positive correlation.
- Blue cells indicate negative correlation.
- White cells indicate no correlation. 
- The darker the colors, the stronger the correlation (positive or negative) between those two variables as indicated by
  cmap = 'coolwarm'.
'''
```
